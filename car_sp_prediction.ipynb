```python
category_map1 = {'Petrol':0, 'Diesel':1, 'CNG':2}
category_map2 = {'Dealer':0, 'Individual':1}
category_map3 = {'Manual':0, 'Automatic':1}

df['Fuel_Type'] = df['Fuel_Type'].map(category_map1)
df['Seller_Type'] = df['Seller_Type'].map(category_map2)
df['Transmission'] = df['Transmission'].map(category_map3)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven']]:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    LL = Q1 - 1.5 * IQR 
    UL = Q3 + 1.5 * IQR 
    df = df[(df[i] >= LL) & (df[i] <= UL)]
    
df
```

```python
x = df.drop(columns=['Car_Name', 'Selling_Price', 'Owner', 'Seller_Type'])
x
```

```python
y = df['Selling_Price']
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
x_train.isnull().sum()
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
residuals = y_test - y_pred

plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_pred, y=residuals, alpha=0.7)
plt.axhline(y=0, color='r', linestyle='--', label='Zero Residual Line')
plt.xlabel('Predicted Sales Price')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Distribution of Residuals')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(y_test - y_pred, bins=30, kde=True)
plt.xlabel('Error')
plt.ylabel('Frequency')
plt.title('Histogram of Prediction Errors')
plt.show()
```

```python
errors = y_test - y_pred

plt.figure(figsize=(10, 6))
plt.errorbar(x=y_test, y=y_pred, yerr=abs(errors), fmt='o', alpha=0.7, ecolor='red')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', label='Ideal Prediction')
plt.xlabel('Actual Sales Price')
plt.ylabel('Predicted Sales Price')
plt.title('Actual vs. Predicted Sales Price with Error Bars')
plt.legend()
plt.show()
```

```python
import scipy.stats as stats

plt.figure(figsize=(10, 6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('Q-Q Plot of Residuals')
plt.show()
```

```python
from scipy import stats

stat, p_value = stats.shapiro(residuals)

print(f'Shapiro-Wilk Test Statistic: {stat}')
print(f'p-value: {p_value}')
```

```python
from sklearn import metrics

MAE = metrics.mean_absolute_error(y_test, y_pred)
MSE = metrics.mean_squared_error(y_test,y_pred)
RMSE = np.sqrt(MSE)
R2 = metrics.r2_score(y_test, y_pred)

print('MAE:', MAE)
print('MSE:', MSE)
print('RMSE:', RMSE)
print(f'R2: {R2:.2f}')
```
